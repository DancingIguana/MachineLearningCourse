## Sección de repaso.



### Primera parte.
Supongamos que tenemos un conjunto de datos que son linealmente separables y queremos realizar la tarea de clasificación binaria (por ejemplo, ¿es perro?).

***¿Regresión logística genera la misma recta que separa los datos en comparación con máquinas de soporte vectorial en $R^n$? ¿cuál es la diferencia entre regresión logística y SVM?***
 
**R:** No va a generar la misma recta, la razón de esto es que ambos algoritmos manejan los datos de manera diferente. En las máquinas de soporte vectorial se quieren encontrar la recta que mejor se ajuste geométricamente a un margen de separación lo más grande posible entre los soportes vectoriales de cada clase. Esto último con el fin de reducir los potenciales errores que puede haber en los datos. Mientras tanto en la regresión logística no estamos buscando que los datos se ajusten de la mejor manera geométrica a la separación marcada por la línea/hiperplano, lo que estamos haciendo es optimizar una función con distintos pesos que permitan optimizar la "probabilidad" de que un punto sea de determinada clase dependiendo de su posición en el plano. Podría decirse en resumen que la regresión logística se apoya en métodos de probabilidad para optimizar los puntos; y las máquinas de soporte vectorial se apoyan en propiedades geométricas de los datos para hacer la separación.

Otro aspecto notable es que la regresión logística considera para el ajuste de su recta todos los puntos del plano. En cambio, la máquina de soporte vectorial solo se enfoca en los puntos más cercanos al hiperplano de separación, los cuales son los vectores de soporte. 

Al considerar esto, concluimos que no van a hacer la misma recta

````python
#Crea un conjunto de puntos con una distribución normal.
centers = [[10,10] , [13,13]]
X, y = make_blobs(n_samples=100, centers=centers, random_state=40)
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
````

````python



def plot_classifier(X,y ,clf = None):
  h = .01 
  x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
  y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                          np.arange(y_min, y_max, h))
  plt.figure()

  if(clf != None):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
  
  
  colors = "bry"
  for i,color in zip(set(y), colors):
      idx = np.where(i == y)
      plt.scatter(X[idx, 0], X[idx, 1], c=color)

  plt.show()
````

````python
plot_classifier(X_train,y_train)
````

````python
clf = LogisticRegression()
clf.fit(X_train, y_train)
plot_classifier(X_train,y_train,clf)
````

````python
clf1 = SVC(kernel = "linear")
clf1.fit(X_train,y_train)
plot_classifier(X_train,y_train,clf1)
````

````python

RL = accuracy_score(clf.predict(X_test) , y_test)
SVM = accuracy_score(clf1.predict(X_test) , y_test)

print("Resultados con regresión logística: %f \n Resultados con SVM: %f" %(RL,SVM))
````



**¿Por qué no generan la misma recta? ¿Qué trata de minimizar cada algoritmo?**

**R:** No generan la misma recta porque están tratando de optimizar distintos parámetros. La forma en que se define cada uno de estos métodos va a llevar a distintos resultados que separarán lo mejor que puedan el conjunto de datos, pero como se enfocan en distintos aspectos, van a resultar en distintas rectas.

Recordemos cuál es el objetivo general de la **máquina de soporte vectorial** con su definición:

Cuando tenemos datos que son linealmente separables, la máquina de soporte vectorial se va a encargar de buscar un hiperplano óptimo que separe las clases. Para ello va a solo escoger unos cuantos puntos del plano para influencia su decisión, estos puntos serán aquellos que están más cerca del hiperplano. Se conocen como vectores de soporte.

Los vectores de soporte son los que definen la posición del hiperplano en un modelo de máquinas de soporte vectorial. El problema a resolver en este método consiste en encontrar cuáles son los vectores de soporte óptimos para separar los datos. Para comenzar, se define un hiperplano como:
$$\textbf{w}^T\cdot \textbf{x} + b = 0$$
Y este tiene dos líneas paralelas que se intersectan con los vectores de soporte, las cuales definen el margen. Estas se representan por:
$$\textbf{w}\cdot \textbf{x} +b = -1$$
$$\textbf{w}\cdot \textbf{x} + b = +1$$

La distancia más corta que hay entre estas últimas dos líneas se le conoce como el margen. Así que lo que buscamos es encontrar un clasificador (hiperplano) con un margen tan grande como sea posible. Si aplicamos fórmulas de distancia para estos márgenes y el hiperplano y hacemos los procedimientos correspondientes, tendremos que la distancia entre estas dos líneas que definen al margen es:
$$dist = 2/||w||$$

Entonces, con el objetivo de maximizar este margen lo que queremos **minimizar** es este valor de $||w||$, que es la norma de nuestro vector, de pesos para la ecuación del hiperplano. Esto siempre tomando en cuenta de que nuestros datos sí son linealmente separables.

Ahora enfoquémonos en lo que trata de optimizar la **regresión logística**.

Lo que tratamos de hacer es considerar la "probabilidad" de que un vector $x_i$, sea de clase una clase/label 1 (siendo las dos clases 1 y -1). Esta probabilidad se representa de la siguiente forma. :

$$P(y = 1 | x_i;\textbf{w}) = \frac{1}{1+e^{-\textbf{w}^T\textbf{x}}}$$ 

En el caso de que quisiéramos encontrar la probabilidad de que sea clase -1, tendríamos:
$$P(y = -1 | x_i;\textbf{w}) = 1 - \frac{1}{1+e^{-\textbf{w}^T\textbf{x}}}$$

A partir de estas funciones se puede llegar a derivar una función que ayuda a obtener las probabilidades de los datos. Esto implica que mientras más lejos se encuentren los datos del hiperplano en el lado correcto de su clase, mejor va siendo la línea que se ajusta. Esto se encuentra maximizando los parámetros de **w** aplicados a esta función de probabilidad. Si quisiéramos aplicar esto mismo minimizando, pues basta con minimizar la probabilidad negativa de los datos con un cambio de signo a la función.

Como podemos ver, ambos métodos están enfocados en resolver aspectos distintos de un conjunto de datos para encontrar la mejor línea. La máquina de soporte vectorial maximiza un margen y solo le importa el valor de sus soportes vectoriales. Por otro lado, la regresión logística quiere maximizar una "probabilidad" y para ello considera a la vez todos los datos que hay en el plano.

### Segunda parte.

***¿Qué sucede cuando los datos no son linealmente separables?***

**R:** Ya no podemos considerar únicamente los puntos más cercanos al hiperplano como los que definirán definitivamente el tamaño del margen. Si no hacemos algo, entonces nuestro hiperplano no podría separar los datos, ya que no encontraría vectores de soporte tales que se maximize un margen.

Lo que se debe añadir son restricciones que permitan suavizar qué tan estricto es el algoritmo para elegir sus soportes vectoriales. Ahora se formarán los márgenes del hiperplano y puede haber puntos vectores de soporte que estén al lado opuesto del margen de su clase correspondiente. Esta penalización está marcada por una variable $C$ cuyo valor podemos ajustar. Si tuviera valor $C = 0$, estaríamos diciendo que ignore totalmente los datos que están al lado opuesto del margen de su clase. Mientras más grande sea el valor de $C$, mayor penalización habrá por tener datos al lado opuesto de su margen correspondiente. El valor de esta penalización también depende de la distancia mínima entre el margen del lado de la clase correspondiente y el dato fuera de lugar.


Por ejemplo, considera la siguiente gráfica:


````python
Image(filename="/content/sample_data/CSVM.png")
````

En el conjunto de datos hay un dato que estorba para hacer la clasificación trazando una línea. La diferencia entre la solución de la izquierda y la de la derecha es la constante que se utiliza para relajar las restricciones. 

**¿Cuál es la diferencia entre ellas?**

**R:** Como ya habíamos comentado en la pregunta anterior, el valor de esta penalización que indica la variable $C$ influencia qué tan permisivo es el margen con los vectores de soporte dentro de él. Mientras mayor sea $C$ más difícilmente estará permitiendo que haya muchos vectores de soporte al lado opuesto de su margen de clase correspondiente. Esto hace que probablemente el margen se vea obligado a ser algo pequeño al compararlo con $C$s más pequeñas. Mientras menor sea $C$, no va a haber tanta penalización por tener vectores de soporte al lado opuesto del margen correspondiente de la clase, pudiendo tener así aún más oportunidad de hacer el margen más grande. Con esto, al tener menos restricción puede que tengamos un margen más amplio pero con muchos más datos por dentro de él.

Esto es visible en el ejemplo de estas dos figuras. En la figura de la izquierda vemos como el valor $C=1$, permite que haya muy pocos vectores de soporte del lado opuesto al margen que le corresponde a la clase. En este caso solo hay un vector de soporte rojo al lado opuesto que permitió. En el caso de $C=0.01$ tenemos un parámetro que es prácticamente 100 veces más pequeño. Esto le indica al algoritmo que penalice muy poco a los puntos que se encuentren al lado opuesto de su margen correspondiente. Por esta misma razón vemos que hay muchos más puntos rojos y verdes al lado opuesto de su margen correspondiente. También es notorio que el margen es más grande que el de $C=1$.

**¡La magia de SVM!**
 
Resulta que la representación actual de los datos no nos permite utilizar las técnicas conocidas. SVM realiza transformaciones cuando el conjunto de datos no es linealmente separable ( incrementa el número de dimensiones para cambiar la representación) por una que  se "ajuste a los datos" utilizando distintos tipos de kernel.
 
Por ejemplo, veamos la figura que se muestra a continuación. En la gráfica de la izquierda no es posible trazar una línea que separe los datos, mientras que en la transformación que produce la gráfica de la derecha, sí se puede hacer.
 
La transformación en este caso es la siguiente: tomando como focos los dos puntos amarillos que rodean al morado, los nuevos puntos se acomodan marcando en el eje horizontal la distancia al foco izquierdo, y en el vertical la distancia al foco derecho, haciendo además una traslación en el plano para separar los puntos de los ejes coordenados. ¿Puedes separarlos ahora utilizando una línea?

Sklearn ofrece una implementación, donde puede ajustar el kernel que deseas utilizar. 
- linear 
- poly (polynomial kernel)
- rbf  (Gaussian Radial Basis)


````python
clf1 = SVC(kernel = "linear")
clf1.fit(X,y)
plot_classifier(X,y,clf1)
````

````python
clf1 = SVC(kernel = "poly")
clf1.fit(X,y)
plot_classifier(X,y,clf1)
````

````python
clf1 = SVC(kernel = "rbf")
clf1.fit(X,y)
plot_classifier(X,y,clf1)
````

Este algoritmo como menciona al inicio, funciona con etiquetas binarias. Ahora supongamos que queremos clasificarlos en puntos azules, verdes y rojos.

**¿cómo transformamos este algoritmo para clasificar entre estas etiquetas?**
(Hint: one vs one)

La solución a esto **en implementación de Python** es muy sencilla. Lo único que tenemos que hacer es ajustar los parámetros de la función de nuestro modelo para la máquina de soporte vectorial de tal forma que podamos hacer clasificación multiclase. Normalmente se tiene por defecto que la clasificación de los datos sea binaria para la función de ```SVC```, pero si le especificamos que siga la estrategia para resolver problemas de clasificación para tres o más clases, entonces obtendremos la clasificación que queremos. Este tipo de esquema de solución en la documentación se llama **one-vs-one** y la forma en que lo llamamos es especificando el siguiente parámetro dentro de nuestra función de```decision_function_shape = 'ovo'```.

### Desarrollo de actividad

Lo primero que vamos hacer, será cargar nuestros datos en un dataframe de **pandas**. Nos enfocamos de momento en los de entrenamiento. En el archivo que tenemos, hay 784 columnas dedicadas a los valores de los pixeles de la imagen y una a la clase a la que pertenecen (que es un número entre 0 y 9). Así que leemos el csv y guardamos en una serie de **pandas** las clases de cada fila y en una tabla los valores de los pixeles.

````python
training_minst = pd.read_csv("./train.csv")
training_labels = training_minst["label"]


training_minst.drop("label", axis = 1, inplace = True)
training_minst
````

````python
training_labels
````

Ya que tenemos preparados nuestros datos de entrenamiento, lo que sigue es definir un modelo de entrenamiento de máquinas de soporte vectorial con la función ```SVC``` de **sklearn**. Estuve probando distintos *kernels* en el parámetro de la función y el que tuvo buen resultado fue el de ```rbf```. Además de eso, otro parámetro importante es el de ```decision_function_shape```. Acorde a la documentación de **sklearn**, cuando queremos resolver problemas de clasificación multiclase se usa el esquema de "one-vs-one". Se menciona que es esencial usar esto con el propósito de hacer clasificación de múltiples clases y como eso es justo lo que buscamos (porque tenemos 10 clases), vamos a usar el parámetro de: ```decision_function_shape = 'ovo'```. Con estas dos especificaciones será suficiente construir nuestro modelo de la máquina de soporte vectorial.

````python
clf1 = SVC(kernel = "rbf", decision_function_shape = 'ovo')
````

Con el modelo definido, lo entrenamos con nuestro conjunto de datos de entrenamiento y sus clases correspondientes.

````python
clf1.fit(training_minst,training_labels)
````

Ya que tenemos nuestro modelo listo, llegó el momento de probarlo con el conjunto de datos de prueba que ofrece la misma página de donde sacamos el conjunto de entrenamiento de MNIST. En este caso el conjunto no cuenta con las clases de cada fila de datos, así que no podremos realizar en este mismo notebook una evaluación del porcentaje de precisión de nuestro modelo. Primero cargamos el conjunto de prueba:

````python
test_minst = pd.read_csv("./test.csv")
test_minst
````

Hacemos la predicción del conjunto de prueba para cada "imagen". Esto nos va a devolver un número entre 0 y 9 que será supuestamente la mejor predicción en relación a la imagen del número trazado a mano.

````python
class_pred = clf1.predict(test_minst)
````

````python
class_pred
````

Como vemos, ya tenemos un arreglo soluciones correspondientes a cada una de nuestras imágenes. Lo último que haremos será evaluar la precisión del modelo. Para ello, primero guardaremos nuestros datos en un archivo .csv llamado ```output.csv```.

````python
predictions = pd.DataFrame({"ImageId": [i+1 for i in range(0,len(class_pred))], "Label": class_pred})
````

````python
predictions.to_csv('output.csv', columns = ["ImageId", "Label"], index = False)
````
