### Manipulación del conjunto de datos.

En el momento de revisar el conjunto de datos con el que estamos trabajando nos encontramos con datos faltantes. Hay muchas formas de poder lidiar con ellos. 


***¿Cuáles se te ocurren?***


Algunos métodos para lidiar con datos faltantes podrían ser: 
- para valores numéricos asignar la media o mediana de todos los valores que sí tengamos.
- Para valores binarios o de categorías escoger aquel valor que sea más predominante en el conjunto de datos. Aunque también es posible otro que se vea conveniente.
- Si tenemos más que suficientes datos y una parte muy pequeña de ellos está incompleta, podemos simplemente borrar las filas que tengan datos incompletos.
- Si una columna está casi siempre vacía, puede ser conveniente eliminarla.
- Pueden añadirse datos de manera aleatoria que sean consistentes con el contenido de la columna.
- Elegir un valor por defecto para todos los datos faltantes.

 
Hemos decidido reemplazar el valor faltante con el número 0. La intuición aquí es que queremos desplazarnos en dirección hacia donde nos orienta el error (este problema se convierte en un problema de optimización y la forma en como trabaja el algoritmo es por ascenso de la gradiente).

La ecuación que ajusta los pesos es la siguiente:

$ W = W + \alpha * error(x_1,...,x_n)[x_1,...,x_n]$
 
Así, al hacer 0 los valores faltantes, el peso en ese atributo quedaría sin alterar.

Notemos además que, si apareciera un dato etiquetado totalmente incompleto, quedaría clasificado de manera neutral dado que la sigmoide evaluada en 0 nos da 0.5


````python
#función para conseguir el conjunto de datos.
def get_data(path:str):
  data = pd.read_csv(path , sep = " " , names = list(range(28)) , header = None)

  #Los datos nulos se encuentran señalados por un signo ?
  data.replace('?' , 0 , inplace=True)

  #Importante! queremos clasificar unicamente en si vive o muere. 
  label = data.loc[: , 22] #Las etiquetas
  label.replace("3" , 2,inplace= True)

  #Tomaremos las primeras 21 columnas.
  data.drop(columns = range(22,28), inplace = True)
  label = label.astype(int)

  return data, label
````

***¿Qué significa el número 3 en label?¿Por qué reemplazamos ese valor por 2 ?***
 
 
Si vemos la documentación de los datos, vemos que el número 3 en el label significa que un caballo fue sacrificado. Como no hacemos distinción entre si un caballo murió o fue sacrificado, simplemente les asignamos a los sacrificados también la etiqueta 2, que significa que murió

````python
#Dirección del conjunto de datos.
Path  = "./content/sample_data/" 
print(Path + "/horse-colic.data")
training_set , label_training_set = get_data(Path + "/horse-colic.data")  
test_set , label_test_set = get_data(Path + "/horse-colic.test" )
````

### Regresión logística con sklearn.

````python
 clf = LogisticRegression(random_state=0)
````

````python
#Ajustar el algoritmo con el conjunto de datos.
clf.fit(training_set , label_training_set)
````

````python
Y_pred = clf.predict(test_set)
````

## Actividad: modificando datos faltantes

### Opción 1: Media de las columnas 

La primera solución que se me ocurre para llenar los datos faltantes, es llenándolos con la media de los valores numéricos de cada columna. Por ejemplo, si una columna tiene media '3.5', entonces todos los valores faltantes de esa columna serán asignados con el valor 3.5. La desvantaja que puedo pensar de este caso es que al hacer esto, puede que estemos dando valores que realmente no existan en la definición de los datos de la columna. por ejemplo si estamos manejando valores discretos y les asignamos una media, pues tendremos un resultado parecido a ellos pero que no pertenezca en su conjunto. De todas formas vamos a intentar obsevar qué tan bien se ajusta al modelo que tenemos.

Primero establecemos la función para leer los datos y cargamos el conjunto de entrenamiento y prueba. Por conveniencia los valores que teníamos asignados previamente como '?' que pasaron a ser 0, los convertiremos en valores ```NaN```.

````python
#función para conseguir el conjunto de datos.
def get_data(path:str):
  data = pd.read_csv(path , sep = " " , names = list(range(28)) , header = None)

  #Los datos nulos se encuentran señalados por un signo ?
  data.replace('?' , 0 , inplace=True)

  #Importante! queremos clasificar unicamente en si vive o muere. 
  label = data.loc[: , 22] #Las etiquetas
  label.replace("3" , 2,inplace= True)

  #Tomaremos las primeras 21 columnas.
  data.drop(columns = range(22,28), inplace = True)
  label = label.astype(int)

  return data, label
````

````python
#Dirección del conjunto de datos.
Path  = "./content/sample_data/" 
training_set , label_training_set = get_data(Path + "/horse-colic.data")  
test_set , label_test_set = get_data(Path + "/horse-colic.test" )
training_set.replace(0,np.nan, inplace=True)
test_set.replace(0,np.nan, inplace = True)
training_set
````

Ya tenemos nuestras dos tablas de datos y columnas de clases listas. Lo que sigue es llenar las celdas vacías con la media correspondiente de cada columna. Esto se puede hacer rápidamente con la función de ```.fillna()``` de **pandas**.

````python
training_set = training_set.fillna(training_set.astype(float).mean())
test_set = test_set.fillna(test_set.astype(float).mean())

training_set = training_set.astype(float)
test_set = test_set.astype(float)

training_set
````

Una columna que pienso puede estar causando ruido en la clasificación es la **2**. La razón de esto, es que si nos fijamos en la documentación de los datos de entrenamiento, estos valores son identificadores, ID. Es decir, cada caballo supuestamente tiene un identificador único y esta no debería ser un dato que influya en si murió o no. No es como que el nombre de un ser vivo afecte su estado de salud física. Por estas razones, vamos a tirar esta columna de los conjuntos de datos.

````python
training_set = training_set.drop(2, axis = 1)
test_set = test_set.drop(2,axis=1)
training_set
````

El paso que sigue es el de cargar la función de ```LogisticRegression()``` para aplicar el modelo de entrenamiento. En este caso le vamos a añadir un nuevo parámetro de ```max_iter=5000```. La razón por la cual hice esto fue debido a que si mantenía el número máximo de iteraciones por defecto, me marcaba advertencias de que el método no alcanzó a converger, por eso aumenté el máximo numero de iteraciones permitidas en el algoritmo. Hacer esto hizo que ya no aparecieran esos problemas y que el método de regresión logística de la librería sirviera.

````python
clf = LogisticRegression(random_state=0, max_iter = 5000)
````

Ya que cargamos la función, lo que sigue es aplicarla sobre el conjunto de entrenamiento con la función ```.fit()``` como se muestra a continuación:

````python
clf.fit(training_set , label_training_set)
````

El siguiente paso es aplicar el modelo de regresión logística entrenado al conjunto de datos de prueba. Esto lo hacemos con ```predict()``` que nos devolverá la lista de predicciones de clase para cada fila de datos.

````python
Y_pred = clf.predict(test_set)
Y_pred
````

Para darnos idea de qué tan bien podría estar podemos comparar esta lista de predicciones con la de las clases reales:

````python
np.array(label_test_set)
````

El último paso es observar qué tan buena fue la precisión del algoritmo de predicción esta vez. Para ello simplemente obtenemos el ```accuracy_score``` comparando los resultados predecidos con los reales

````python
accuracy_score(label_test_set , Y_pred)
````

Como podemos observar, este método terminó arrojando resultados satisfactorios. Al usar la media de cada columna y tirar las columnas de ID, es notorio como hubo un incremento de precisión en las predicciones. En el caso de llenar todos los datos con 0 teníamos una precisión de aproximadamente 70%. Pero ahora en esta ocasión aumentó un 6% la precisión. A partir de esto podemos concluir que añadir la media a los datos nulos y quitar las columnas de ID en este caso es mejor solución que llenar con 0s los datos faltantes.

### Opción 2: valores aleatorios de columnas

El siguiente caso que probaremos será elegir un valor aleatorio no nulo de la columna y colocarlo en los elementos vacíos del dataframe. Es decir, imaginemos que nuestra columna tiene el conjunto de datos \[1,2,3\]. En esta columna tenemos dos celdas vacías $a$ y $b$. Con esta solución lo que haremos es llenar la celda $a$ con un valor aleatorio del conjunto de datos presentes en la columna, por ejemplo $a = 3$; repetimos el proceso con $b$ eligiendo otro número aleatorio de la columna que resultaría en algo como $b = 1$.

Usando esta idea general, repitamos los mismos pasos que con las implementaciones que hicimos anteriormente.

Primero cargamos nuestros conjuntos de datos, en este caso no hay problema al representar los datos faltantes con 0.

````python
#Dirección del conjunto de datos.
Path  = "./content/sample_data/" 
training_set , label_training_set = get_data(Path + "/horse-colic.data")  
test_set , label_test_set = get_data(Path + "/horse-colic.test" )
#test_set.replace(0, np.nan, inplace = True)
#training_set.replace(0, np.nan, inplace = True)
training_set
````

Ahora reemplazamos los valores faltantes de cada columna por un elemento aleatorio de la misma. Para este caso vamos a estar rellenando con los valores que hay en las columnas de *training_test*. Para cada celda se elegirá un valor aleatorio diferente para así variar más los datos. Esto lo vamos a hacer con ayuda de la librería ```random```. También hay que tener cuidado de que no agarre un elemento ```NaN``` del conjunto de elementos, por esta misma razón los vamos a quitar de las listas de donde vamos a escoger valores.

````python
import random
for column in training_set:
    for element in range(0,len(training_set[column])):
        if(training_set[column][element] == 0):
            training_set[column][element] = random.choice(list(set(training_set[column].values)- {np.nan}))
    for element in range(0, len(test_set[column])):
        if(test_set[column][element] == 0):
            test_set[column][element] = random.choice(list(set(training_set[column].values) - {np.nan}))

training_set = training_set.astype(float)
test_set = test_set.astype(float)

training_set
````

Por las mismas razones que en la sección de la **opción 1**, vamos a eliminar de nuestro conjunto de datos la columna de identificadores: 2. Esto lo hacemos en nuestro conjunto de entrenamiento y en el de prueba.

````python
training_set = training_set.drop(2,axis = 1)
test_set = test_set.drop(2, axis = 1)
````

Los siguientes pasos son los mismos que hemos estado siguiendo en las dos pruebas de modelo anteriores: cargar el modelo, ajustar el modelo a los datos de entrenamiento, predecir con el modelo generado los datos del conjunto prueba y acabar con la precisión final de las predicciones generadas comparándolas con las clasificaciones reales.

Recordando todo esto, simplemente repetimos los pasos y tenemos que:

````python
clf = LogisticRegression(random_state=0,max_iter = 5000)
````

````python
clf.fit(training_set , label_training_set)
````

````python
Y_pred = clf.predict(test_set)
Y_pred
````

````python
accuracy_score(label_test_set , Y_pred)
````

Notemos que obtuvimos la suerte de que nuestro modelo de entrenamiento haya tenido también buenos resultados (77% de precisión) como las otras dos propuestas. Algo notable de esta solución es que puede que no haya mucho control sobre los resultados que arrojen las predicciones debido a la aleatoriedad con la que manejamos nuestros datos faltantes. En este caso el porcentaje fue mayor que las otras dos opciones que tuvieron 70% y 76% de precisión. Aunque esto parece benéfico, también podría ser posible que en una nueva ejecución del programa, nos arroje una precisión menor que las otras dos soluciones propuestas en este laboratorio. Al final, sigue siendo una opción válida, ya que nunca sabremos realmente cómo eran los datos faltantes, solo podemos ir lanzando estimaciones que nunca serán la verdad total.

### Opción 3: Usar la mediana de las columnas

El siguiente procedimiento a considerar es usar esta vez la mediana de las columnas. En este caso hay una ventaja por encima del método de la media y es que si ahora tenemos valores discretos de una columna esta vez sí estaremos asumiendo únicamente valores que existen dentro de la misma y ya no habrá problemas de asignar algún dato que no debía existir a los valores vacíos.

Repetimos el mismo procedimiento que en las otras tres ocasiones, primero cargamos nuestros datos. En este caso sí es conveniente reemplazar los datos faltantes '?' que pasan a 0 por un ```NaN```.

````python
#Dirección del conjunto de datos.
Path  = "./content/sample_data/" 
training_set , label_training_set = get_data(Path + "/horse-colic.data")  
test_set , label_test_set = get_data(Path + "/horse-colic.test" )
test_set.replace(0, np.nan, inplace = True)
training_set.replace(0, np.nan, inplace = True)
training_set
````

Lo que sigue es asignarle a los datos faltantes el valor de la mediana de las columnas. En este caso asignaremos la mediana de la columna del conjunto entrenamiento a los dos conjuntos para que no varíen la forma en que decidimos entrenar y rellenar los datos. Esto lo hacemos como se muestra a continuación:

````python
training_set = training_set.fillna(training_set.astype(float).median())
test_set = test_set.fillna(test_set.astype(float).median())

training_set = training_set.astype(float)
test_set = test_set.astype(float)

training_set
````

También tiramos en este caso las columnas de ID:

````python
training_set = training_set.drop(2,axis = 1)
test_set = test_set.drop(2, axis = 1)
````

Repetimos el procedimiento que ya conocemos para aplicar regresión logística con **sklearn** que ya hemos hecho anteriormente. Nos fijamos al final en el porcentaje de precisión:

````python
clf = LogisticRegression(random_state=0,max_iter = 5000)
````

````python
clf.fit(training_set , label_training_set)
````

````python
Y_pred = clf.predict(test_set)
Y_pred
````

````python
accuracy_score(label_test_set , Y_pred)
````

Podemos observar que usar la mediana de los datos también pareció una buena solución para llenar datos faltantes. Como vemos, el porcentaje de precisión en este caso es del 75%, mayor que el caso de llenar con 0s, pero menor que el caso de usar la media y valores aleatorios (por golpe de suerte este último). Viendo que sigue dando buenos resultados, es una solución aceptable al momento de lidiar con datos faltantes.

### Opción 4: Interpolación con pandas

Otra forma de lidiar con datos faltantes es con una función que viene en la librería de **pandas** para interpolar los datos faltantes. Esta es la de ```.interpolate()```. Repetimos el proceso de cargar los datos que ya hemos visto anteriormente, en este caso sí conviene reemplazar los 0s de datos faltants por ```NaN```.

````python
#Dirección del conjunto de datos.
Path  = "./content/sample_data/" 
training_set , label_training_set = get_data(Path + "/horse-colic.data")  
test_set , label_test_set = get_data(Path + "/horse-colic.test" )
test_set.replace(0, np.nan, inplace = True)
training_set.replace(0, np.nan, inplace = True)
training_set
````

Ahora aplicamos el método de interpolación que viene en la librería de **pandas**. En este caso es necesario usar el parámetro de *limit_direction = "both"*, si no lo usamos obtendremos datos ```NaN``` en las columnas que tienen elementos vacíos en sus primeras filas.

````python
training_set = training_set.astype(float).interpolate(limit_direction="both")
test_set = test_set.astype(float).interpolate(limit_direction="both")

training_set = training_set.astype(float)
test_set = test_set.astype(float)

training_set
````

También vamos a tirar las columnas innecesarias de la ID del caballo en esta ocasión:

````python
training_set = training_set.drop(2,axis = 1)
test_set = test_set.drop(2, axis = 1)
````

Ahora aplicamos de la misma forma el algoritmo de regresión logística de **sklearn** sobre estos datos y checamos el porcentaje de precisión de la predicción.

````python
clf = LogisticRegression(random_state=0,max_iter = 5000)
````

````python
clf.fit(training_set , label_training_set)
````

````python
Y_pred = clf.predict(test_set)
Y_pred
````

````python
accuracy_score(label_test_set , Y_pred)
````

Vemos que usar la función de interpolación que viene en pandas para lidiar con datos faltantes arroja buenos resultados para entrenar al modelo de predicción. En este caso resultó tener exactamente la misma precisión que usar la mediana, ambas siendo del 50%. Concluimos que esto también podría ser una solución válida para lidiar con datos faltantes en este modelo.

### Opción 5: elegir el elemento más frecuente de la columna

Como última solución para lidiar con datos faltantes vamos a probar elegir el elemento más frecuente de cada columna. Por ejemplo, si en una columna el elemento que más aparece es el 1, entonces vamos a llenar todos los datos faltantes de esa columna con 1.

Repetimos el procedimiento anteriormente visto, dejando como  ```NaN``` los valores faltantes:

````python
#Dirección del conjunto de datos.
Path  = "./content/sample_data/" 
training_set , label_training_set = get_data(Path + "/horse-colic.data")  
test_set , label_test_set = get_data(Path + "/horse-colic.test" )
test_set.replace(0, np.nan, inplace = True)
training_set.replace(0, np.nan, inplace = True)
training_set
````

Ahora le añadimos el valor más frecuente de cada columna a los valores ```NaN``` como se muestra a continuación:

````python
for column in training_set:    
    training_set[column].replace(np.nan, (training_set[column].value_counts().idxmax()), inplace = True)
    test_set[column].replace(np.nan, (training_set[column].value_counts().idxmax()), inplace = True)
training_set
````

También vamos a quitar las columnas de ID en los dos conjuntos de datos (entrenamiento y prueba):

````python
training_set = training_set.drop(2,axis = 1)
test_set = test_set.drop(2, axis = 1)
````

Repetimos el proceso para entrenar el modelo de predicción y así predecir los datos del conjunto de entrenamiento que ya hemos hecho anteriormente:

````python
clf = LogisticRegression(random_state=0,max_iter = 5000)
````

````python
clf.fit(training_set , label_training_set)
````

````python
Y_pred = clf.predict(test_set)
Y_pred
````

````python
accuracy_score(label_test_set , Y_pred)
````

Como podemos observar, obtuvimos buenos resultados también en este caso. Al igual que en el uso de la media, la precisión resultó ser de 76%, la cual coincidió con la que había al usar la media de la columna para los datos faltantes. Al observar que sí tenemos buenos resultados a partir de usar el elemento más frecuente de cada columna, concluimos que esta también puede llegar a ser una buena solución para este tipo de problemas.

## Problema de clasificación con tres clases:

### Idea para solucionar el problema

Para un problema de clasificación con tres clases seguiría haciendo el mismo procedimiento que el planteado en este laboratorio. La única diferencia sería la adición de unos parámetros en la función de regresión logística que tiene **sklearn**. Lo primero que se haría es cargar los datos, especificando cúal de las tres clases  ('rojo', 'azul' y 'verde') tiene cada uno y  sus dos atributos respectivos. En el caso de que haya datos faltantes podríamos aplicar cualquier método que veamos conveniente, como eliminar filas, reemplazar por 0, usar promedios o medianas, etc. Algo que también haría sería aplicar el **encoder** mencionado durante la explicación de este laboratorio, así no tendría problemas con los valores de los datos.

Una vez que se tiene todo el conjunto de datos, lo que seguiría es aplicar las funciones de **sklearn**. Primero llamamos a la función de ```LogisticRegression()```,  cuya documentación claramente nos indica que este algoritmo de entrenamiento es capaz de resolver problemas de clasificación "multiclase". Esto es justamente lo que buscamos, ya que tradicionalmente pensamos en una clasificación binaria, pero la modificación de esta función permite hacerlo con 3 o más clases también. Para ello un parámetro que se le puede añadir a la función es el de *multi_class = 'multinomial'*, o también el *multi_class = ovr*. Ambos siendo métodos implementados y diseñados para lidiar con la clasificación de múltiples clases en regresión logística.

Una vez que se defina esta función, la aplicamos a los datos que tuvimos del inicio y hacemos lo mismo que con las clasificaciones de este laboratorio: ajustar el algoritmo al conjunto de datos con ```.fit()``` sobre el conjunto de entrenamiento. Finalmente se hace la predicción con ```predict()``` sobre los datos cuya clase queremos estimar y lo comparamos con los resultados verdaderos para comprobar la eficacia con ```accuracy_score```.
